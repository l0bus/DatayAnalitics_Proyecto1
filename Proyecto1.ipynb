**Ejercicio N°1:**

Escribir un programa que pregunte al usuario o genere las ventas
de los últimos 5 días, y muestre por pantalla una serie (Dataframe)
con los datos de las ventas por producto (opcional): Nro de
unidades y precio unitario, indexada por los días, antes y después
de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

```python
# Importamos librerias a usar
import pandas as pd
import numpy as np
```

```python
# Aca pondremos las variables a usar y sus valores iniciales.
productos = []
precios = []
fechas = []
cantidad = []
columnas = ["Producto", "Precio/Unitario","Cantidad","Precio Total",
            "Precio Total con Descuento"]
mercaderia = [['arroz',2],['azucar',2.5],['leche',3],['gaseosa',9.5],
              ['mantequilla',6.4],['queso',24],['sublime',2.2],
              ['naranja',2.3],['platano',0.5],['Aceite',2.4],['Detergente',7]]
```

```python
# Definimos las funciones a utilizar
def DataManual():
  '''
  Funcion DataManual encargada de ingresar los valores de las ventas
  manualmente, Ingresando la Fecha, Producto, Valor Unitario, Cantidad
  vendida.
  Ejemplo.
  21082022
  Arroz
  2.5
  5
  '''
  global producto, precio
  dato=''
  while(True):
    print("""Desea Salir escriba: Y, Si desea continuar Ingresando Datos presione cualquier letra""")
    opcion = input()
    if (opcion == 'y') or (opcion == 'Y'):
      break
    else:
      print("Ingresar Fecha de Venta(YYYY-MM-DD): ")
      dato0 = input()
      fechas.append(dato0)
      print("Ingresar Producto Vendido: ")
      dato1 = input()
      productos.append(dato1)
      print("Ingresar Precio Unitario de ",dato1,": ")
      dato2 = float(input())
      precios.append(dato2)
      print("Ingresar Cantidad Vendida de ",dato1,": ")
      dato3 = float(input())
      cantidad.append(dato3)
```

```python
#Iniciamos el programa, teniendo 2 opciones, ingreso manual o generacion automatica
print("""Buenos dias.
Si usted va a ingresar manualmente sus ventas seleccione 1.
Si usted va a generar una lista de ventas seleccione 2.
Seleccione la opcion correcta""")
EntradaDatos = input()
if (EntradaDatos == '1'):
  DataManual()
elif (EntradaDatos == '2'):
  numero_ventas=np.random.randint(1,100) #Primero tenemos un valor del numero de ventas
  #Asignamos fechas entre hoy y 5 dias atras
  fecha = np.arange(np.datetime64(np.datetime64('today','D') - np.timedelta64(5, 'D')),np.datetime64('today','D')) 
  fechas = np.random.choice(fecha,numero_ventas) #Generamos fechas aleatorias para el numero de ventas
  for i in range(0,numero_ventas): #Generamos las ventas de productos aleatorias y sus precios
    temp = np.random.randint(len(mercaderia))
    productos.append(mercaderia[temp][0])
    precios.append(mercaderia[temp][1])
  cantidad = np.random.randint(1,10,numero_ventas) #Generamos la cantidad de productos que se vendieron
else:
  exit
```

```python
#Generamos el Dataframe Solicitado
df_ventas = pd.DataFrame(None,fechas,columnas)
df_ventas['Producto'] = productos
df_ventas['Precio/Unitario'] = precios
df_ventas['Cantidad'] = cantidad
df_ventas['Precio Total'] = df_ventas['Precio/Unitario']*df_ventas['Cantidad']
df_ventas['Precio Total con Descuento'] = df_ventas['Precio Total']-(df_ventas['Precio Total']*0.05)
df_ventas
```

```python
# Mostramos el dataframe solicitado agrupado por producto
df_product = df_ventas.groupby(['Producto']).sum()
print(df_product[['Cantidad','Precio Total','Precio Total con Descuento']])
```

mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.
El resultado final debe ser una serie con los estadísticos.

```python
#Generamos un diccionario clientes
clientes = {
    "Nombre": ["Carlos Berroti","Ernesto Rodriguez","Daniel Marroqui","Juana Samanez","Sofia Benavidez","Eduado Marazas","Enrique Mostajo",
               "Rafel Bernedo", "Jorge Ramirez","Julio Esquiche"],
    }
clientes["Ingresos"] = np.random.randint(2000,15000,len(clientes['Nombre']))
#Ingresamos en diccionario y lo convertimos en Dataframe y mostramos los valores solicitados
df_ingresos = pd.DataFrame(clientes,None,['Nombre','Ingresos'])
df_ingresos
```

```python
df_stadistic = pd.DataFrame(None,["Max. Ingreso","Min. Ingreso","Promedio Ingreso"],["Nombre",'Ingresos'])
df_stadistic.loc["Max. Ingreso"] = df_ingresos.max()
df_stadistic.loc["Min. Ingreso"] = df_ingresos.min()
df_stadistic.loc["Promedio Ingreso"] = df_ingresos.median()
df_stadistic['Ingresos']
```

Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

***

***
Enero 30500 22000
***
Febrero 35600 23450
***
Marzo 28300 18100

Abril 33900 35700

***
A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas&gt;= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto&gt;Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
columns = ["Mes","Ventas","Gastos"]
Mes = ["Enero","Febrero","Marzo","Abril","Mayo"]
Ventas = [30500,35600,28300,33900,42500]
Gastos = [22000,23450,18100,35700,32450]
df_datos = pd.DataFrame(None,None,columns)
df_datos["Mes"] = Mes
df_datos["Ventas"] = Ventas
df_datos["Gastos"] = Gastos
df_datos
df_datos["C1 (G/V)*100"] = 100*df_datos["Gastos"]/df_datos["Ventas"]
df_datos.loc[df_datos["Ventas"] >= 30000, "C2(V>=30000)"] = "Meta Superada"
df_datos.loc[df_datos["Ventas"] < 30000, "C2(V>=30000)"] = "Meta No Superada"
df_datos.loc[df_datos["Ventas"] > df_datos["Gastos"], "C3(V>G)"] = df_datos["Ventas"] - df_datos["Gastos"]
df_datos.loc[df_datos["Ventas"] < df_datos["Gastos"], "C3(V>G)"] = "-"
df_datos
```

```python
Suma_Ventas1 = df_datos.iloc[-3::]['Ventas'].sum()
print("La suma de las ventas de los ultimos 3 meses = ",Suma_Ventas1)
Suma_Ventas2 = df_datos.iloc[1:4:]['Ventas'].sum()
print("La suma de las ventas de los meses de febrero a abril = ",Suma_Ventas2)
```
