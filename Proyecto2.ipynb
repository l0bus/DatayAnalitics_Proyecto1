```python
# creamos funcion 
def fpais(parm1):
   vpais = parm1;
   if vpais.upper() == 'USA':
      rcode = 1
   elif vpais.upper() == 'EUROPE':
      rcode = 2
   elif vpais.upper() == 'JAPAN':
      rcode = 3
   else: 
      rcode = 0
   return rcode
```

```python
from google.colab import files
df_archivo = pd.read_csv("/content/autos.csv",sep=";")
df_archivo
```

```python
# Verificamos si existen datos nulos por columna
df_archivo.isnull().sum()
```

```python
#Empezamos el tratamiento de nuestros datos.
columnas=['Millas_Por_Galon','Caballo_De_Fuerza']
df_archivo[columnas] = df_archivo[columnas].fillna(df_archivo[columnas].mean())
```

```python
# Verificamos si existen datos nulos por columna
df_archivo.isnull().sum()
```

```python
# Log10(var1/var2)
df_archivo["columna_1"] = df_archivo.apply(lambda row: mt.log10(row["Cilindros"]/row["Caballo_De_Fuerza"]), axis=1)

# Sqrt(var1)exp(var2)/200
df_archivo["columna_2"] = df_archivo.apply(lambda row: mt.sqrt(row["Cilindros"])*mt.exp(row["Caballo_De_Fuerza"])/200, axis=1)

# Si var 1 > var 2 → 5, caso contrario 3
df_archivo["columna_3"] = df_archivo.apply(lambda row: 5 if row["Cilindros"] > row["Caballo_De_Fuerza"] else 3, axis=1)

# 1/logn(var1/var2)*100
df_archivo["columna_4"] = df_archivo.apply(lambda row: 1/mt.log(row["Cilindros"]/row["Caballo_De_Fuerza"])*100, axis=1)

# var2**2/var1
df_archivo["columna_5"] = df_archivo.apply(lambda row: row["Cilindros"]**2/row["Caballo_De_Fuerza"], axis=1)

df_archivo
```

```python
#Generamos Variables Cualitativas

vmed = np.mean(df_archivo["Millas_Por_Galon"])
vstd = np.std(df_archivo["Millas_Por_Galon"])

# Rango Inicial
df_archivo["RangoInicial"] = df_archivo.apply(lambda row: (vmed - 1*vstd), axis=1)

# Si es mayor a la “media + 1*desviación estándar” → Alto
df_archivo["columna_1"] = df_archivo.apply(lambda row: "Alto" if row["Millas_Por_Galon"] >= (vmed + 1*vstd) else np.nan, axis=1)

# Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
df_archivo["columna_2"] = df_archivo.apply(lambda row: "Medio" if row["Millas_Por_Galon"] > (vmed - 1*vstd) and row["Millas_Por_Galon"] < (vmed + 1*vstd) else np.nan, axis=1)

# Si es menor a la “media - 1*desviación estándar” → Bajo
df_archivo["columna_3"] = df_archivo.apply(lambda row: "Bajo" if row["Millas_Por_Galon"] <= (vmed - 1*vstd) else np.nan, axis=1)

# Rango Final
df_archivo["RangoFinal"] = df_archivo.apply(lambda row: (vmed + 1*vstd), axis=1)

# mostrar las columnas que se estan evaluando
df_archivo.loc[:, ["RangoInicial", "Millas_Por_Galon", "RangoFinal", "columna_1", "columna_2", "columna_3"]]

```

```python
# Creamos una funcion por grafica

def fcajasybigotes():
    print("Cajas y bigotes")    
    print("------------------------------------")
    fig, ax = plt.subplots(figsize=(6, 3.84))  

    sns.boxplot(
                x     = 'Cilindros',  # variable cualitativa
                y     = 'Millas_Por_Galon',  # variable cuantitativa
                data  = df_archivo,   # dataframe
                ax    = ax
            )
    ax.set_title('Millas_Por_Galon por Cilindros');
def fbarradelamedia():
    mean_df = df_archivo.mean()
    print("Barras de la media")
    print("------------------------------------")
    plt.figure(figsize=(7,5))
    sns.countplot(df_archivo['Pais_Origen'], order=df_archivo['Pais_Origen'].value_counts().index)
def fhistograma():
    print("Histograma con Curva")
    print("------------------------------------") 
    sns.displot(data=df_archivo, x="Pais_Origen", kde=True)

# creamos funcion para los graficos
def fgraficos(parm1):
   vopcion = int(parm1);
   if vopcion == 1:      
      fcajasybigotes()
      rcode = 'Opción: 1'
   elif vopcion == 2:      
      fbarradelamedia()
      rcode = 'Opción: 2'
   elif vopcion == 3:      
      fhistograma()
      rcode = 'Opción: 3'
   elif vopcion == 4:
      fcajasybigotes()      
      fbarradelamedia()      
      fhistograma()
      rcode = 'Opción: 4'
   else: 
      rcode = 'Valor Invalido'
   return rcode
```
